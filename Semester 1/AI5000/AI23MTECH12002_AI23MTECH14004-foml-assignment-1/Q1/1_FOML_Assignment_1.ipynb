Regression Modelling

````python
# as per the question , we can use existing packages like scikit-learn for linear regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
````

````python
# Splitting the data into training and testing sets (80% train, 20% test)
X = encoded_data.drop(columns=["Date", "Time to finish food (in mins )(including waiting time)"])
y = encoded_data["Time to finish food (in mins )(including waiting time)"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=43)
````

*Linear Regression*

````python
# Add a column of ones to the feature matrix for the bias term
X_train_bias = np.c_[np.ones((X_train.shape[0], 1)), X_train.values]
X_test_bias = np.c_[np.ones((X_test.shape[0], 1)), X_test.values]
````

````python
# Computing the weights using the pseudo-inverse
pseudo_inverse = np.linalg.pinv(X_train_bias.T.dot(X_train_bias))
weights = pseudo_inverse.dot(X_train_bias.T).dot(y_train)
````

````python
# Predict on the test set
linear_preds_numpy = X_test_bias.dot(weights)
````

````python
# Evaluate the model
linear_mae_numpy = np.mean(np.abs(y_test - linear_preds_numpy))
linear_rmse_numpy = np.sqrt(np.mean((y_test - linear_preds_numpy)**2))
linear_mae_numpy, linear_rmse_numpy
````

**Poisson Regression**

````python
class PoissonRegressionMLE(object):
    def __init__(self, lr=0.01, num_iter=1000):
        self.lr = lr
        self.num_iter = num_iter

    def fit(self, X, y):
        # Adding a column of ones for the intercept
        X = np.column_stack((np.ones(X.shape[0]), X))
        self.coefficients = np.zeros(X.shape[1])
        for _ in range(self.num_iter):
            y_pred = np.exp(X @ self.coefficients)
            gradient = X.T @ (y - y_pred)
            self.coefficients += self.lr * gradient / len(X)

    def predict(self, X):
        X = np.column_stack((np.ones(X.shape[0]), X))
        return np.exp(X @ self.coefficients)
````

````python
poisson_reg_mle = PoissonRegressionMLE()
poisson_reg_mle.fit(X_train.values, y_train.values)
preds = poisson_reg_mle.predict(X_test.values)
mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))
mae, rmse
````

Exponential Regression

````python
class ExponentialRegressionMLE(object):
    def fit(self, X, y):
        # Taking the natural logarithm of the response variable
        y_log = np.log(y)
        # Adding a column of ones for the intercept
        X = np.column_stack((np.ones(X.shape[0]), X))
        # Computing the MLE estimates for the coefficients
        self.coefficients = np.linalg.inv(X.T @ X) @ X.T @ y_log

    def predict(self, X):
        X = np.column_stack((np.ones(X.shape[0]), X))
        return np.exp(X @ self.coefficients)
````

````python
exp_reg_mle = ExponentialRegressionMLE()
exp_reg_mle.fit(X_train.values, y_train.values)
preds = exp_reg_mle.predict(X_test.values)
mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))
mae, rmse
````
